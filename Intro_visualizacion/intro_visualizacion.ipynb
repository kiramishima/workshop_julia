# Graficando datos de un DataFrame

```julia
using DataFrames
using CSV
```

```julia
df = CSV.read("../intro_dataframes/medals_by_country_2016.csv", DataFrame)
```

```julia
plot(df.country, df.Gold)
```

# Customizando graficos

```julia
plot(df.country, df.Gold,
    xrotation=90,
    palette=palette([:purple, :green], 7),
    size=(800, 800)
)
xlabel!("PAÍSES")
ylabel!("TOTAL")
title!("MEDALLAS DE ORO POR PAÍS")

# Colorschemes: https://docs.juliaplots.org/latest/generated/colorschemes/
```

# Scatterplot

```julia
plot(df.country, df.Gold,
    seriestype=:scatter,
    xrotation = 90,
    palette=palette([:purple, :green], 7),
    size=(800, 800)
)
xlabel!("PAÍSES")
ylabel!("TOTAL")
title!("MEDALLAS DE ORO POR PAÍS")
```

```julia
scatter(df.country, df.Bronze,
    xrotation = 90,
    size=(800, 800)
)
xlabel!("PAÍSES")
ylabel!("TOTAL")
title!("MEDALLAS DE BRONCE POR PAÍS")
```

# Lineas

```julia
scatter(df.country, df.Silver,
    seriestype=:line,
    xrotation = 90,
    size=(800, 800)
)
xlabel!("PAÍSES")
ylabel!("TOTAL")
title!("MEDALLAS DE PLATA POR PAÍS")
```

# Barras

```julia
plot(df.country, df.Gold,
    seriestype=:bar,
    xrotation = 90,
    size=(800, 800)
)
xlabel!("PAÍSES")
ylabel!("TOTAL")
title!("MEDALLAS DE ORO POR PAÍS")
```

```julia
bar(df.country, df.Gold,
    xrotation = 90,
    size=(800, 800)
)
xlabel!("PAÍSES")
ylabel!("TOTAL")
title!("MEDALLAS DE ORO POR PAÍS")
```

```julia
bar(df.country, df.Silver, 
    permute=(:x,:y),
)
ylabel!("PAÍSES")
xlabel!("TOTAL")
title!("MEDALLAS DE ORO POR PAÍS")
```

# Stackbar

```julia
groupedbar([df.Gold df.Silver df.Bronze],
        bar_position = :stack,
        bar_width=0.7,
        xticks=(1:10, df.country),
        xrotation=90,
        leg=:right,
        xlabel="PAÍSES",
        ylabel="TOTAL",
        title="MEDALLAS POR PAÍS",
        group = repeat(["Gold", "Silver", "Bronze"], inner=length(df.country)),
        label=["Gold" "Silver" "Bronze"]
)
```

# Guardando graficos

```julia
groupedbar([df.Gold df.Silver df.Bronze],
        bar_position = :stack,
        bar_width=0.7,
        xticks=(1:10, df.country),
        xrotation=90,
        leg=:right,
        size=(800, 800),
        xlabel="PAÍSES",
        ylabel="TOTAL",
        title="MEDALLAS POR PAÍS",
        group = repeat(["Gold", "Silver", "Bronze"], inner=length(df.country)),
        label=["Gold" "Silver" "Bronze"]
)

savefig("total_medallas.png")
```
